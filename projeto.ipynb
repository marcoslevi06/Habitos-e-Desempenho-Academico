```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
CAMINHO_BASE_DE_DADOS = "arquivos_base/student_habits_performance.csv"
df = pd.read_csv(f"{CAMINHO_BASE_DE_DADOS}")
# print(df.head(5))
# print(df.info())
df
```

```python
# RENOMEANDO AS COLUNAS PARA PORTUGUÊS.

colunas_pt = {
    "student_id": "id_estudante",
    "age": "idade",
    "gender": "gênero",
    "study_hours_per_day": "horas_estudo_por_dia",
    "social_media_hours": "horas_redes_sociais",
    "netflix_hours": "horas_netflix",
    "part_time_job": "emprego_meio_periodo",
    "attendance_percentage": "percentual_presença",
    "sleep_hours": "horas_sono",
    "diet_quality": "qualidade_dieta",
    "exercise_frequency": "frequência_exercícios",
    "parental_education_level": "escolaridade_pais",
    "internet_quality": "qualidade_internet",
    "mental_health_rating": "avaliacao_saude_mental",
    "extracurricular_participation": "participação_extracurricular",
    "exam_score": "nota_exame"
}
df.rename(columns=colunas_pt, inplace=True)
df
```

```python
# CRIANDO UM HEATMAP (mapa de calor) DOS NOSSOS DADOS.
colunas_desejadas = ["horas_estudo_por_dia", "horas_redes_sociais", "horas_netflix", 
                     "percentual_presença", "horas_sono", "avaliacao_saude_mental",
                     "nota_exame"] 

# Reformatando o DataFrame Pandas utilizando o método .corr() para criar uma correlação entre os dados.
plt.title("Correlação entre hábitos e nota final.")
df_heatmap = sns.heatmap(data=df[colunas_desejadas].corr(), annot=True, cmap="YlGnBu")
```

##### Alunos que mais estudam têm um desempenho melhor?

```python
# Gráfico de dispersão com linha d eregressão.
# x = "horas_estudo_por_dia" / y = "nota_exame"

sns.lmplot(data=df,x="horas_estudo_por_dia", y="nota_exame", col="gênero", height=9)
plt.title("Horas de estudo -> Notas do Exame?")
# plt.show()
```

##### Comparando as notas de quem estuda mais de 5h por dia e quem estuda menos de 2h por dia.

```python
filtro_mais_de_5h = df["horas_estudo_por_dia"] > 5
filtro_menos_de_2h = df["horas_estudo_por_dia"] < 2

grupo_mais_de_5h = df[filtro_mais_de_5h]["horas_estudo_por_dia"]
grupo_menos_de_2h = df[filtro_menos_de_2h]["horas_estudo_por_dia"]

print(f"Média das notas das pessoas que estudam MAIS que 5h: ", round(grupo_mais_de_5h.mean(), 4))
print(f"Média das notas das pessoas que estudam MENOS que 2h: ", round(grupo_menos_de_2h.mean(), 4))
```

##### O tempo gasto em redes sociais afeta o desempenho dos alunos?

```python
# Redes sociais: distribuição geral (Histrograma).

eixo_x = "horas_redes_sociais"

sns.histplot(df, x=eixo_x, color="skyblue", kde=True, bins=30)
plt.title("Distribuição de tempo em redes socias.")
plt.show()
```

##### Quebrando a amostra dos dados pela quantidade de horas nas redes sociais para comparar com as notas do exame.

```python
df["horas_redes_sociais_corte"] = pd.cut(
    df["horas_redes_sociais"],
    bins=[0, 1, 2, 3, 4, 5, 6],
    # labels=["0-2", "2h-4h", "4h-6h"]
)

sns.boxplot(x="horas_redes_sociais_corte", y="nota_exame", data=df)
plt.title("Notas por tempo em redes sociais.")
plt.show()
```

##### Verificando a correlação entre alunos mais saudáveis terem melhor desempenho.

```python
for coluna in ["frequência_exercícios", "avaliacao_saude_mental", "qualidade_dieta"]:
    sns.boxplot(x=coluna, y="nota_exame", data=df, color="#d953bd")
    plt.title(f"{coluna.title()} relativa a Nota do Exame")
    plt.show()
```
